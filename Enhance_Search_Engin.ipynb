```python
import zipfile
import io

count = 0

def decode_method(binary_data):
    global count
    # Decompress the binary data using the zipfile module
    # print(count, end=" ")
    count += 1
    with io.BytesIO(binary_data) as f:
        with zipfile.ZipFile(f, 'r') as zip_file:
            # Assuming there's only one file in the ZIP archive
            subtitle_content = zip_file.read(zip_file.namelist()[0])

    # Now 'subtitle_content' should contain the extracted subtitle content
    return subtitle_content.decode('latin-1')  # Assuming the content is UTF-8 encoded text
```

```python
df['file_content'] = df['content'].apply(decode_method)
```

```python
import re

def clean_subtitles(text):

     text = text.lower()
    
    
     text = re.sub(r'\d+\s*\n\d{2}:\d{2}:\d{2},\d{3}\s*-->.*\n', '', text)

   # 2. Remove HTML tags
     text = re.sub(r'<.*?>', '', text)

   # 3. Remove special characters (keeping basic punctuation)
     text = re.sub(r'[^\w\s,.\'!?]', '', text)

   # 4. Remove extra white spaces (new lines, spaces, etc.)
     text = re.sub(r'\s+', ' ', text).strip()

    
     return text
```

```python
df["cleaned_text"] = df["file_content"].apply(clean_subtitles)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain.schema import Document

# Initialize the text splitter
text_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " ", ""],
    chunk_size=2000,
    chunk_overlap=50,
)

# Function to split text into chunks
def split_text(text):
    doc = Document(page_content=text)
    return text_splitter.split_documents([doc])


# Apply the text splitter to each document in the series
df["chunks"] = df["cleaned_text"].apply(split_text)

```

```python
import chromadb
client = chromadb.Client()
```

```python
collection = client.create_collection(name="Database")
#collection = client.get_collection(name="Database")
```

```python
collection.count()
```

```python
collection.peek()
```

```python
# Convert df["chunks"] to a list of strings
chunks_list = df["chunks"].tolist()

# Ensure all elements are strings
chunks_list = [str(chunk) for chunk in chunks_list]

# Add to ChromaDB collection
collection.add(
    documents=chunks_list,
    ids=[f"id{i}" for i in range(len(chunks_list))]
)
```

```python
# Perform a simple query to check if embeddings exist
results = collection.query(
    query_texts=["script info title english"],
    n_results=1
)

# Check if results are returned
if results['documents']:
    print("Embeddings successfully added.")
else:
    print("No embeddings found.")
```
